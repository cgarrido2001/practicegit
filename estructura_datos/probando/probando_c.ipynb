# Tipos de Datos

```python
%%file Tiposdedatos.c

#include<stdio.h>

int main(){
    char a = 'e'; //tamano = 1 byte, rango: 0..255
    short b = -15; //tamano = 2 bytes, rango : -128...127
    int c = 1024; //tamano  = 2bytes, rango: -32768...32767
    unsigned int d = 128; //tamano= 2bytes, rango: 0...65535 solo enteros positivos o 0
    long e = 123456; //tamano= 4bytes, para printear se nesecita long int -> %li
    float f = 3.1416; //tamano= 4bytes
    double g = 12345.123456; //tamano = 8bytes -> long float -> %lf
    printf("El elemento es: %lf", g);
    
    return 0;
}
```

```python
%%bash

gcc Tiposdedatos.c -g -o Tiposdedatos
./Tiposdedatos
```

# Entrada y salida de datos

### Para almacenar datos se utiliza el scanf("%tipo_dato",&x[direccion de memoria de la variable])

el scanf solo lee hasta que encuentre un espacio por lo que si se digita un nombre y un apellido solo guarda el nombre.

para arreglar esto se puede usar gets(x)

```python
%%file input_output.c

#include<stdio.h>

int main(){
    char x[50];
    
    printf("dime tu nombre: ");
    gets(x);
    
    printf("tu nombre es: %s", x);
    
    
    return 0;
}
```

```python
%%bash

gcc input_output.c -g -o input_output
```

### Ejercicio hipotenusa

```python
%%file hipotenusa.c

#include<stdio.h>
#include<math.h>

int main(){
    
    float a,b,c;
    
    printf("Ingresa el valor del cateto 1: ");
    scanf("%f", &a);
    printf("Ingresa el valor del cateto 2: ");
    scanf("%f", &b);
    
    c = sqrt(pow(a,2)+pow(b,2));
    
    printf("la hipotenusa es: %.2f\n", c);
    

    return 0;
}
```

### Condicionales

&& es and y || es or

```python
%%file sentencia_if.c

#include<stdio.h>

int main(){
    int n1,n2;
    
    printf("Digita 2 numeros: ");
    scanf("%i %i",&n1, &n2);
    
    if(n1 % n2 == 0){
        printf("El numero %i es divisible por %i", n1, n2);
    }
    else{
        printf("los numeros no son divisibles");
    }

    return 0;
}
```

### Sentencia while

```python
%%file bucle_while.c

//mostrar los 10 primeros numeros en pantalla

#include<stdio.h>

int main(){
    
    int i = 1;
    while(i<=10){
        printf("%i ",i);
        i++;
    
    }

    return 0;
}
```

```python
!gcc bucle_while.c -g -o bucle_while
```
