```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10; #Puertos
puertos = [i for i in range(n)];
rutas = [(i,j) for i in puertos for j in puertos if i!=j];

puertos, rutas
```

### Crear coordenandas aleatorias para cada puerto, en un mapa de 100x100[mÂ²]

```python
random = np.random;
random.seed(10);

coordenadaX = random.rand(n)*100;
coordenadaY = random.rand(n)*100;

print(f"Las Coordendas X para todos los puertos: \n {coordenadaX}\n");
print(f"Las Coordendas Y para todos los puertos: \n {coordenadaY}");
```

### Graficamos los puertos en un scatterplot

```python
x = coordenadaX;
y = coordenadaY;

plt.figure(figsize=(12,6))
plt.scatter(coordenadaX, coordenadaY)
plt.title("Mapa de coordenadas de los puertos")
plt.xlabel("Coordenadas X")
plt.ylabel("Coordenadas Y")


#Mostramos las coordenadas de cada puerto en el mapa
s = []
for i in range(len(coordenadaX)):
    s_temp = []
    s_temp.append("%.1f"%coordenadaX[i])#coordenandaX
    s_temp.append("%.1f"%coordenadaY[i])#coordenandaY
    s.append(s_temp)#agrego [coordenadaX, coordenadaY] a la lista s

#Mostrar en el mapa
for i in range(len(s)):
    plt.annotate(str(s[i]), xy=(coordenadaX[i], coordenadaY[i]), xytext=(coordenadaX[i]-4, coordenadaY[i]-4),
                 color="purple")

#Ahora identificamos cada puerto con un numero

for i in range(len(s)):
    plt.annotate(str(i),xy=(coordenadaX[i], coordenadaY[i]), xytext=(coordenadaX[i], coordenadaY[i]+2), color='green')
plt.show()
```
